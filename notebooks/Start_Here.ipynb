# Welcome to PSDS 2300

## Data Visualization

__Synopsis:__ 
This course covers the fundamental concepts of current visualization concepts and technologies. 
Unlike many data visualization courses, this course focuses on the principles of visualization design and the grammar of graphics. 
These principles are then implemented in popular contemporary visualization technologies accessed through Python and R API. 
Students will develop an advanced knowledge of the appropriate selection, modeling, and evaluation of data visualizations. 

__Course Design:__ 
This is a 3 credit hour / 5 day course.
This course is delivered using a blend of in classroom instructrion and self-paced online classroom learning. 
The instructor kicks off the course in the NGA classroom on day-one; 
followed by three additional days over a two-week period that are used for self-paced, online classroom learning; 
and then a final day with the instructor in the NGA classroom to recap material, lead contextual discussions, and finalize the course. 
Self-paced online learning activities utilize iPython and IRKernel within JupyterHub. 


All PSDS courses are taught at the graduate level. 
Consequently, the amount of effort required will vary based on individuals prior educational background and work experience. 
As a result, some students may need to spend additional time and effort outside of the scheduled instruction days to successfully complete the required course material.
